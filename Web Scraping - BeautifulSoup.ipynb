```python
#Book Details - first 2 pages

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']
base_url="http://books.toscrape.com/catalogue/"
bookDetails=[]
for i in allPages:
    response=requests.get(i)
    data=BeautifulSoup(response.text, 'html.parser')
    book=data.find_all(class_='product_pod')
    for j in book:
        books_url=base_url+j.h3.a['href']
        r=requests.get(books_url)
        data=BeautifulSoup(r.text, 'html.parser')
        title=data.h1.string
        link=books_url
        price=data.find(class_='price_color').string
        price=float(re.search('[\d.]+',price).group())
        qty=data.find(class_='instock availability').contents[-1].strip()
        qty=int(float(re.search('\d+',qty).group()))
        book_data=[title,link,price,qty]
        bookDetails.append(book_data)
        
df=pd.DataFrame(bookDetails, columns=column_names)

for i in bookDetails:
    print(*i)
```

```python
#fetching all the details ofthe book and storing in a csv file

import requests
from bs4 import BeautifulSoup
import re
import pandas as pd

#storing all the links of next page in "base_url" variable

base_url='http://books.toscrape.com/catalogue/'
current_url='http://books.toscrape.com/catalogue/page-1.html'
url_list=[]
r=requests.get(current_url)
while r.status_code==200:
    url_list.append(current_url)
    data=BeautifulSoup(r.text, 'html.parser')
    next1=data.find(class_='next')
    if next1 is None:
        break
    next_url=base_url+next1.a['href']
    current_url=next_url
    r=requests.get(current_url)
    
#storing details of all the books in "bookDetails" variable

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']
base_url="http://books.toscrape.com/catalogue/"
bookDetails=[]
for i in url_list:
    response=requests.get(i)
    data=BeautifulSoup(response.text, 'html.parser')
    book=data.find_all(class_='product_pod')
    for j in book:
        books_url=base_url+j.h3.a['href']
        r=requests.get(books_url)
        data=BeautifulSoup(r.text, 'html.parser')
        title=data.h1.string
        link=books_url
        price=data.find(class_='price_color').string
                #extracting numbers from the string and type casting to float
        price=float(re.search('[\d.]+',price).group())
        qty=data.find(class_='instock availability').contents[-1].strip()
                #extracting numbers from the string and type casting to float
        qty=int(float(re.search('\d+',qty).group()))
        book_data=[title,link,price,qty]
        bookDetails.append(book_data)
        
#creating DataFrame to store the all the data of 'bookDetails'

df=pd.DataFrame(bookDetails, columns=column_names)
print(df.head)

#storing all the data in csv file
df.to_csv('books.toscrpe', index=False)
```
